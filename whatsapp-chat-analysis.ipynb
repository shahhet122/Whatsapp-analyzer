```python
import re
import pandas as pd
```

```python
f = open('chat.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?[AP]M\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
date_pt = r'\u202f'
def filtering_date(string):
    return re.sub(date_pt, '', string)
```

```python
new_dates = [filtering_date(date) for date in dates]
new_dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date':new_dates})
df
df['message_date'] = pd.to_datetime(new_dates, format='%m/%d/%y, %I:%M%p - ')
df.rename(columns={'message_date':'date'}, inplace=True)
df.head(35)
```

```python
df.shape
```

```python
#separating username and messages from the user_message col.
users = []
messages=[]
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_noti')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df.head(30)
```

```python
#separating day,month and year from the date col.

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df
```
